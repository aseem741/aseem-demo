```python
# LCM Calculator
# Write a Python program to find the least common multiple (LCM) of two positive integers.
from math import gcd
num1 = int(input("enter the 1st value :"))
num2 = int (input("enter 2nd value :"))
lcm = (num1*num2) // gcd(num1,num2)
print ("LCM =",lcm)
```
